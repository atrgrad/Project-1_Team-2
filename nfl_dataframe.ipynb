```python
import pandas as pd
import os
from pathlib import Path
from dotenv import load_dotenv
%matplotlib inline
```

```python
superbowl_df = Path ("../Project-1_Team-2/superbowl_data.csv")

superbowl = pd.read_csv(
    superbowl_df, index_col="Year")
superbowl.iloc[:,0] = superbowl.iloc[:,0].shift(1)
superbowl.head()
```

```python
raw_data = pd.DataFrame([])
years_start = 1996
years_end = 2022
for year in range(years_start,years_end):
        url = f'https://www.pro-football-reference.com/years/{year}/preseason_odds.htm'
        df = pd.read_html(url)[0]
        df['year'] = year
        raw_data = raw_data.append(df, ignore_index=True, sort=False)
raw_data 
```
